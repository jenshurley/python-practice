```python
xvals = np.arange(0, 10, .10)
#because it's going to go on the x axis, xvals

plt.plot(xvals, np.sin(xvals), label='sin')
plt.plot(xvals, np.cos(xvals), label='cosine')
#an easy way to get data to understand how plots work
#applying sin to xvals data

plt.xlim(0,10)
plt.ylim(-1,1)
#if you don't want to show all the values - limit the values (try changing 10 to 5)

plt.xlabel('Sin/Cos of x',size=12,fontweight='semibold')
plt.ylabel('Punction eval of X',size=12,fontweight='semibold')

plt.title("Example Chart of Cos/Sin Functions", size=14, fontweight='bold')
#plt.legend(loc=(0.64,0.1))
plt.legend(loc='best')
```

```python
d = [11,12,13,14,15,16,17]

t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]

plt.plot(d,t0,label='Tokyo')
plt.plot(d,t1,label='Hawaii')
plt.plot(d,t2,label='Hong Kong')

plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12,fontweight='semibold')
plt.ylabel('Temperature (C)',size=12,fontweight='semibold')

plt.grid(True,linewidth=0.5,color='#aaaaaa',linestyle='-')

plt.title("Daily temperature of 3 cities in the second week of December", size=14, fontweight='bold')

plt.legend(loc=(0.64,0.1))
plt.show()
```

```python
# Making multiple subplots
nums = np.arange(0, 10, .10)

plt.subplot(1,2,1) #rows, cols, (position - the first chart or the 2nd...) plot number
plt.plot(nums, np.cos(nums))

plt.subplot(1,2,2) #rows, cols, plot nbr
plt.plot(nums, np.sin(nums))
```

```python
# Making multiple subplots
nums = np.arange(0, 10, .10)

plt.subplot(2,1,1) #rows, cols, plot nbr
plt.plot(nums, np.cos(nums))
plt.ylabel('cos')

plt.subplot(2,1,2) #rows, cols, plot nbr
plt.plot(nums, np.sin(nums))
plt.ylabel('sin')
plt.xlabel('x')
```

```python
fig, axes = plt.subplots(4, 3);
#this will return two variables - unpacking a tuple here

for ax in axes.flatten():
    ax.plot(nums, np.sin(nums))
```

```python
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tanh(nums), label='tan')
plt.plot(nums, np.arcsin(nums), label='arcsin')
plt.legend(loc='best', ncol=2, title='trig functions', fancybox=True)
```

```python
tips.sex.value_counts()
```

```python
# Create a dict of all the colors
def color_dict_sex(sex):
    if sex == 'Male':
        return 'red'
    else:
        return 'green'
    
tips['sex_color'] = tips.sex.map(color_dict_sex)
    
```

```python
scatter = plt.figure()
axes1 = scatter.add_subplot(1,1,1)
axes1.scatter(tips.total_bill, tips.tip, color = tips.sex_color.tolist())

axes1.set_title('Relationship between Tip and Bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip Amount')

```

```python
# Women Tips
wt = tips.query('sex == "Male"').tip.tolist()
mt = tips.query('sex == "Female"').tip.tolist()

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([wt, mt], labels=['Female', 'Male']);
axes1.set_ylim(0,7)


axes1.set_title('BoxPlot of Tips by Gender')
axes1.set_xlabel('Gender')
axes1.set_ylabel('Tip Amount')
```

```python
# Scatter tips by size of the party and gender
scatter = plt.figure()
axes1 = scatter.add_subplot(1,1,1)
axes1.scatter(x=tips.total_bill, 
              y=tips.tip, s=tips.size / 7, c=tips.sex_color, alpha=.5)
axes1.set_title('Tips by Total party size and gender')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip Amount')
```
